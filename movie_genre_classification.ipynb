```python
import pandas as pd

train = pd.read_csv(
    "train_data.txt",
    sep=" ::: ",
    engine="python",
    names=["ID", "Title", "Genre", "Plot"]
)

test = pd.read_csv(
    "test_data.txt",
    sep=" ::: ",
    engine="python",
    names=["ID", "Title", "Plot"]
)

solution = pd.read_csv(
    "test_data_solution.txt",
    sep=" ::: ",
    engine="python",
    names=["ID", "Title", "Genre", "Plot"]
)

print("Train Dataset Shape:", train.shape)
print("Test Dataset Shape:", test.shape)

train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=5000
)

X_train = vectorizer.fit_transform(train["Plot"])
X_test = vectorizer.transform(test["Plot"])

y_train = train["Genre"]
y_test = solution["Genre"]

print("TF-IDF conversion completed!")
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
predictions = model.predict(X_test)

print("Prediction completed!")
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)

print("\nClassification Report:\n")
print(classification_report(y_test, predictions))
```

```python
sample_plot = [
    "A young superhero fights dangerous aliens to save the Earth."
]

sample_vector = vectorizer.transform(sample_plot)

predicted_genre = model.predict(sample_vector)

print("Predicted Genre:", predicted_genre[0])
```

```python
output = test.copy()

output["Predicted_Genre"] = predictions

output.to_csv("movie_genre_predictions.csv", index=False)

print("Predictions saved successfully!")
```
